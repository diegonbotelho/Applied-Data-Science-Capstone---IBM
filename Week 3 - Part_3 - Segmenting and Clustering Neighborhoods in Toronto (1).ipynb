# Part I - Create Dataframe

```python
# Importing required libraries
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import requests # library to handle requests
from urllib.request import urlopen

# website scraping libraries and packages in Python from BeautifulSoup 
from bs4 import BeautifulSoup 

print('Libraries imported.')
```

```python
# Scrap data from wikipedia page & Create dataframe
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_website = urlopen(url) 
soup = BeautifulSoup(wiki_website, 'html.parser')
```

```python
# reading the table and creating array to hold the data from table
table_toronto = soup.find_all('table')

postalcodes = []
boroughs = []
neighborhoods = []

for table in table_toronto:
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all('td')
        if len(cells) == 3:
            postalcode = cells[0]
            postalcodes.append(postalcode.text.strip())
            
            borough = cells[1]
            boroughs.append(borough.text.strip())
            
            neighborhood = cells[2]
            neighborhoods.append(neighborhood.text.strip())
            
# reading the table into a pandas dataframe
df_toronto = pd.DataFrame(postalcodes, columns = ['Postal Code'])
df_toronto['Borough'] = boroughs
df_toronto['Neighborhood'] = neighborhoods

# Ignoring cells with a borough with 'Not assigned'
df_toronto = df_toronto[df_toronto['Borough'] != 'Not assigned'].reset_index(drop = True)

# Separating neighborhoods with a comma
df_toronto = df_toronto.apply(lambda x: x.str.replace('/',','))

# If a cell has a borough but a not assingned neighborhood, then the neigborhood will be the same as the borough
df_toronto['Neighborhood'] = df_toronto[['Neighborhood', 'Borough']].apply(lambda x: x[0] if x[0] else x[1], axis=1)

print(df_toronto.shape)
df_toronto.head()
```

# Part II - Get the latitude and the longitude coordinates of each neighborhood

```python
# Reading CSV file
geo_data = pd.read_csv('https://cocl.us/Geospatial_data')
geo_data.head()
```

```python
# Creating a new dataframe df_toronto and geo_data merged
df = pd.merge(df_toronto, geo_data, on = 'Postal Code', how = 'left')
# Number of rowns and columns of the dataframe
print(df.shape)
df
```

```python
# Get the top 100 venues that are in Toronto's Neighborhoods within a radius of 500 meters
LIMIT = 100 
def getNearbyVenues(names, latitudes, longitudes, radius = 500):
    
    venues_list = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Creating a new datframe called toronto_venues
toronto_venues = getNearbyVenues(names=df_T['Neighborhood'],
                                   latitudes=df_T['Latitude'],
                                   longitudes=df_T['Longitude']
                                  )
print('\nDone!')
```

```python
# Checking the size of the resulting dataframe
print(toronto_venues.shape)
toronto_venues.head()
```

```python
# Checking how many venues were returned for each neighborhood
toronto_venues.groupby('Neighborhood').count()
```

```python
# Analyze Each Neighborhood

# One hot econding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix = "", prefix_sep = "")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 
mid = toronto_onehot['Neighborhood']
toronto_onehot.drop(labels = ['Neighborhood'], axis = 1, inplace = True)
toronto_onehot.insert(0, 'Neighborhood', mid)

print(toronto_onehot.shape)
toronto_onehot.head()
```

```python
# Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.shape
```

```python
# Let's create the new dataframe and display the top 10 venues for each neighborhood
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## 2) Cluster Neighborhoods

```python
# Set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# Run k-means clustering
kmeans = KMeans(n_clusters = kclusters, random_state = 0).fit(toronto_grouped_clustering)

# Check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# Add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = df_T

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on = 'Neighborhood')

toronto_merged.head()
```

```python
# create map
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
